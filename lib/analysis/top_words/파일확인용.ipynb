```python
import pickle
import sys
```

```python
import os
import re
```

```python
mod = sys.modules[__name__]
```

```python
def call_file_list(path) : 
    """
        path : 파일 경로
    """
    from os import listdir
    path_dir = path
    file_list = listdir(path_dir)
    return file_list
```

```python
folder_list = call_file_list('../top_words')[:-2]
```

```python
folder_list
```

```python
i = 0
for folder in folder_list : 
    file_list = call_file_list('./%s'%folder) 
    try : 
        for file in file_list : 
            i +=1
            my = re.compile('(\d{4})[-](\d{2})[-](\d{2})([_][#])(\w*)')
            mytxt = my.search(file)
            
            with open('./{}/{}'.format(folder, file), 'rb') as f : 
                setattr(mod, 'anal_{}_{}'.format(mytxt.group(2)+mytxt.group(3),mytxt.group(5)), pickle.load(f))
            print('anal_{}_{}'.format(mytxt.group(2)+mytxt.group(3),mytxt.group(5)))
    except Exception as e :
        print(e)
```
