```python
ko = KomoranPy()
```

```python
ko.set_user_dictionary('./user_dictionary.txt')
```

```python
with open('../../data/twitch_live_chat/#zilioner/2018-12-05_#zilioner.log') as f : 
    text = f.readlines()
```

```python
text
```

```python
import re
my = re.compile('\[([0-9:]*)\] <(\S*[ ]*\S*)> (\w.*)')
word_list = []
for line in text:
    mytext = my.search(line)
    if mytext :
        word_list.append(mytext.group(3))
word_list
```

```python
tokens_ko = []
for wd in word_list : 
    for i in range(len(ko.pos(wd))) : 
        tokens_ko.append(ko.pos(wd)[i][0])
```

```python
tokens_ko
```

```python
tk = nltk.Text(tokens_ko)
```

```python
data=  tk.vocab().most_common(300)
data
```
