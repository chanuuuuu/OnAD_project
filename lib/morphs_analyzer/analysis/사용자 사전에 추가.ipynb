```python
def call_file_list(path) : 
    """
        path : 파일 경로
    """
    from os import listdir
    path_dir = path
    file_list = listdir(path_dir)
    return file_list
```

```python
def cut_filename(filename, num = 3) : 
    import re
    my = re.compile('([\d]*[-][\d]*[-][\d]*)([_]*)([#\w.]*)([.][\w.]*)')
    txt = my.search(filename)
    mytxt = txt.group(num)
    return mytxt
```
