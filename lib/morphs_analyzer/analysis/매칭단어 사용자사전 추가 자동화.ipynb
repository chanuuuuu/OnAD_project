```python
def call_file_list(path) : 
    """
        path : 파일 경로
    """
    from os import listdir
    path_dir = path
    file_list = listdir(path_dir)
    return file_list
```

```python
def cut_filename(filename, num = 3) : 
    import re
    my = re.compile('([\d]*[-][\d]*[-][\d]*)([_]*)([#\w.]*)([.][\w.]*)([.][\w.]*)')
    txt = my.search(filename)
    mytxt = txt.group(num)
    return mytxt
```

```python
def remove_word(x):
#     line = iter(list(df['morphs'][df['origin']== x]))
#     new_line = next(line)
    tmp = list(df['morphs'][df['origin']== x])[0]
    series_tolist = ast.literal_eval(tmp)
    for wd in series_tolist:
        if wd in x : 
            target_index = x.index(wd)
            remove_word = x[target_index: target_index + len(wd)]
            x = x.replace(remove_word, "")
    return x.strip()
```
