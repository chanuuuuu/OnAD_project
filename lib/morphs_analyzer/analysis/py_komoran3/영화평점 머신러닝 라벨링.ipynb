```python
def call_file_list(path) : 
    """
        path : 파일 경로
    """
    from os import listdir
    path_dir = path
    file_list = listdir(path_dir)
    return file_list
```

```python
file_list = call_file_list('../../data/movie140reviewcorpus-master/raw')
file_list
```

```python
with open('../../data/movie140reviewcorpus-master/raw/10001-드라마-멜로-로맨스.txt') as f : 
    text = f.readlines()
```

```python
text
```

```python
text_tuple = []
for line in text : 
    try : 
        text_tuple.append(ast.literal_eval(line))
    except Exception as e : 
        print(e, '\t',text.index(line))
```

```python
text_tuple
```

```python
df = pd.DataFrame(
    {'rating' : [text_tuple[i][2]for i in range(len(text_tuple))],
     'review' : [text_tuple[i][-1]for i in range(len(text_tuple))]
    }
)
```

```python
df.head()
```

```python
def divide(x) : 
    if x >= 8 : return 'POS'
    elif x <= 5 : return 'NEG'
    else : return 'NEUT'
```

```python
df['emo'] = df['rating'].apply(lambda x : divide(x))
```
