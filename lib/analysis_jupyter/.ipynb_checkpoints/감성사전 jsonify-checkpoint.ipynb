```python
# 단어를 설명하는 문장을 형태소 분석하여 하나의 문장으로 다시 합치기 위해 한 문장을 리스트로 저장 후 그 목록을 더 큰 리스트에 저장
# 갓 : 신이라는 뜻이다 -> 신이라는 뜻이다 -> 신 , 이라는, 뜻, 이다 -> [신, 이라는, 뜻, 이다]
w_list = []
word_list_morph = []
for i in range(len(word_list_list)) : 
    for wd in word_list_list[i] : 
        for mp in api.analyze(wd) :
            for m in mp.morphs :
                w_list.append(m.lex+'/'+m.tag)
        word_list_morph.append(w_list)
        w_list = []
word_list_morph
```

```python
# 긍정단어
pos_pol_word = open("./KnuSentiLex/pos_pol_word.txt", "r", encoding='utf-8').readlines()
```

```python
pos_pol_word = pos_pol_word[19:]
```

```python
pos_pol_word
```

```python
pos_pol_word_dict = {}
```

```python
for i in range(len(pos_pol_word)) :
    for word in api.analyze(pos_pol_word[i]):
        morphs_str = ' + '.join([(m.lex + '/' + m.tag) for m in word.morphs])
        pos_pol_word_dict[morphs_str] = 'POS'
```

```python
pos_pol_word_dict
```

```python
# 부정단어
neg_pol_word = open("../../data/sentiment_dictionary/KnuSentiLex/neg_pol_word.txt", "r").readlines()
```

```python
neg_pol_word = neg_pol_word[19:]
```

```python
neg_pol_word_dict = {}
```

```python
for i in range(len(neg_pol_word)) :
    for word in api.analyze(neg_pol_word[i]):
        morphs_str = ' + '.join([(m.lex + '/' + m.tag) for m in word.morphs])
        neg_pol_word_dict[morphs_str] = 'NEG'
```

```python
neg_pol_word_dict
```

```python
pol_word = {}
```

```python
pol_word['POS'] = pos_pol_word_dict
pol_word['NEG'] = neg_pol_word_dict
```
