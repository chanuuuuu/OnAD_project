```python
import khaiii
import pandas as pd
```

```python
api = khaiii.KhaiiiApi()
api.open()
```

```python
# 감성사전 불러오기
pol = pd.read_csv('../../data/sentiment_dictionary/lexicon/polarity.csv')
```

```python
pol.head()
```

```python
# 감성사전을 담을 딕셔너리 생성
sentiment_dict = {}
sentiment_dict
```

```python
# 감성사전을 딕셔너리 안에 키는 ngram으로 값은 max.value로 저장
for i in range(len(pol)) : 
    sentiment_dict[pol['ngram'][i]] = pol['max.value'][i]
sentiment_dict
```

```python
# 채팅로그 불러오기
text = open('../../data/chat_logs/2018-12-04_yapyap30-Copy1.log', encoding='utf-8').readlines()
```

```python
text
```

```python
# 채팅로그 전처리
import re
my = re.compile('\[([0-9:]*)\] <(\S*[ ]*\S*)> (\S.*)')
word_list = []
for line in text:
    mytext = my.search(line)
    if mytext :
        word_list.append(mytext.group(3))
word_list
```

```python
print('0',my.search(text[2]).group(0),
'1',my.search(text[2]).group(1),
'2',my.search(text[2]).group(2),
'3',my.search(text[2]).group(3)
)
```

```python
# 리스트로 된 채팅로그를 공백으로 조인하여 문자열로 변환
word_ = ' '.join(word_list)
word_
```

```python
# 형태소 분석하여 morphs_list에 리스트로 저장
morphs_list = []
for word in api.analyze(word_):
    morphs_str = ';'.join([(m.lex+'/'+m.tag) for m in word.morphs])
    morphs_list.append(morphs_str)
```

```python
morphs_list
```

```python
# 감성분석하여 점수를 담을 리스트 생성
score_list = []
```

```python
# morphs_list 안에 있는 형태소를 감성사전과 비교하여 점수 매칭하여 score_list에 담기. 키가 없으면 nan 저장
for mph in morphs_list : 
    try :
        score = sentiment_dict[mph]
        score_list.append(score)
    except : 
        score = 'NaN'
        score_list.append(score)
```

```python
score_list
```

```python
# Nan 제거
for i in range(len(score_list)) : 
    try :
        score_list.remove('NaN')
    except : 
        print('제거완료')
```

```python
# none 제거
for i in range(len(score_list)) : 
    try :
        score_list.remove('None')
    except : 
        print('제거완료')
```

```python
score_list
```

```python
score_list_norm = []
for i in range(len(score_list)) : 
    if score_list[i] == 'POS' : 
        score_list_norm.append(1)
    elif score_list[i] == 'NEG' :
        score_list_norm.append(-1)
    else : score_list_norm.append(0)
    
```

```python
score_sum = 0
for i in range(len(score_list_norm)) : 
    score_sum += score_list_norm[i]
score_sum
```
