```python
import khaiii
```

```python
api = khaiii.KhaiiiApi()
api.open()
```

```python
def open_twitch_chat(data, encoding) : 
    chat_text = open(data, encoding=encoding).readlines()
    return chat_text
```

```python
chat_text = open_twitch_chat('../../data/chat_logs/2018-12-04_yapyap30-Copy1.log','utf-8')
```

```python
def filter_twitch_chat(chat_text, num) : 
    import re
    text = chat_text
    my = re.compile('\[([0-9:]*)\] <(\S*[ ]*\S*)> (\S.*)')

    word_list = list()
    for line in text:
        mytext = my.search(line)
        if mytext :
            word_list.append(mytext.group(num))
    return word_list
```

```python
word_list = filter_twitch_chat(chat_text, 3)
```

```python
def filter_twitch_chat_join(word_list) : 
    word_list = word_list
    joined_word_list = ' '.join(word_list)
    return joined_word_list
```

```python
joined_chat = filter_twitch_chat_join(word_list)
```

```python
morphs_str_list = []
```

```python
for word in api.analyze(joined_chat) : 
    morphs_str = ' + '.join([(m.lex + '/' + m.tag) for m in word.morphs])
    morphs_str_list.append(morphs_str)
morphs_str_list
```
