```python
# 감성사전 불러오기
pol = pd.read_csv('../../data/sentiment_dictionary/lexicon/polarity.csv')
```

```python
pol.head()
```

```python
# 감성사전을 담을 딕셔너리 생성
sentiment_dict = {}
sentiment_dict
```

```python
# 감성사전을 딕셔너리 안에 키는 ngram으로 값은 max.value로 저장
for i in range(len(pol)) : 
    sentiment_dict[pol['ngram'][i]] = pol['max.value'][i]
sentiment_dict
```

```python
# 감성분석하여 점수를 담을 리스트 생성
score_list = []
```

```python
# morphs_list 안에 있는 형태소를 감성사전과 비교하여 점수 매칭하여 score_list에 담기. 키가 없으면 nan 저장
for mph in morphs_list : 
    try :
        score = sentiment_dict[mph]
        score_list.append(score)
    except : 
        score = 'NaN'
        score_list.append(score)
```
