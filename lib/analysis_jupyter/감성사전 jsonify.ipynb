```python
import pandas as pd
```

```python
# 감성사전 불러오기
pol = pd.read_csv('../../data/sentiment_dictionary/lexicon/polarity.csv')
```

```python
pol.head()
```

```python
# 감성사전을 담을 딕셔너리 생성
sentiment_dict = {}
sentiment_dict
```

```python
# 감성사전을 딕셔너리 안에 키는 ngram으로 값은 max.value로 저장
for i in range(len(pol)) : 
    sentiment_dict[pol['ngram'][i]] = pol['max.value'][i]
sentiment_dict
```

```python
# json으로 저장
import json
```

```python
with open('./corpus_sentinment.json', 'w', encoding="utf-8") as make_file:
    json.dump(sentiment_dict, fp=make_file, ensure_ascii=False, indent='\t')
```
