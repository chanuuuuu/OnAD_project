```python
import pickle
```

```python
with open('./#beyou0728/matching_2018-12-04_#beyou0728.pickle', 'rb') as f : 
    text = pickle.load(f)
text
```

```python
list1 = []
for i, word in enumerate(text) : 
    split_text = text[i][1].split(' ')
#     '{}\tNNP'.format(split_text)
    for j in range(len(split_text)) : 
        list1.append(split_text[j])
list1
```

```python
list1 = set(list1)
```

```python
list1 = list(list1)
```

```python
list1
```

```python
for word in list1 : 
    print('{}\tNNP'.format(word))
```

```python
from konlpy.tag import Komoran
ko = Komoran()
```

```python
for word in list1 : 
    p = ko.nouns(word)
    print(p)
```

```python
for word in list1 : 
    p = ko.morphs(word)
    print(p)
```
